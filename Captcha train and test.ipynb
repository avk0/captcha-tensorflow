```python
import os
from PIL import Image
import numpy as np
import random

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
```

```python
def img_to_black_and_white(in_path, out_path, w=190):
    with Image.open(in_path) as img:
        for x in range(img.size[1]):
            for y in range(img.size[0]):
                pix = np.array(img.getpixel((y, x)))
                if sum(pix) > w*3:
                    img.putpixel((y, x), (255, 255, 255))
                else:
                    img.putpixel((y, x), (0, 0, 0))
        img.save(out_path)

    
def dir_to_bw(in_dir, out_dir, number=None, w=190):
    os.makedirs(os.path.dirname(out_dir), exist_ok=True)
    if number == None: 
        for captcha in os.listdir(in_dir):
            img_to_black_and_white(in_dir + captcha, out_dir + captcha, w)
    else:
        for captcha in os.listdir(in_dir)[:number]:
            img_to_black_and_white(in_dir + captcha, out_dir + captcha, w)
            
            
def cut_to_5(in_path, out_dir):
    with Image.open(in_path) as img:
        wid = img.size[0]
        hei = img.size[1]
        #print(img.size)
        dw = int(wid/5)
        for piece in range(5):
            temp_img = img.copy().crop((1+piece * dw, 0, (piece + 1) * dw - 1 , hei))
            #print(temp_img.size)
            name = os.path.basename(in_path)[piece]    
            full_name = os.path.join(out_dir, name + os.path.basename(in_path)[5:])
            safe_save(full_name, temp_img)
            #temp_img.save(os.path.join(out_dir, name + os.path.basename(in_path)[5:]))

            
def path_cut_to_5(in_dir):
    for f in os.listdir(in_dir):
        f_path = in_dir + f
        cut_to_5(f_path, in_dir)
        os.remove(f_path)
        
        
def safe_save(file, image, add=0):
    original_file = file
    split = file.split(".")
    part1 = ".".join(split[:-1]) + "_" + str(add)
    file = ".".join([part1, split[-1]])
    #print(file)
    if not os.path.isfile(file):
        #print(file)
        image.save(file)
    else:
        add += 1
        safe_save(original_file, image, add=add)
        

def train_test_split(in_path, ratio = 0.2):
    os.makedirs(os.path.join(in_path, 'test'), exist_ok=True)
    os.makedirs(os.path.join(in_path, 'train'), exist_ok=True)
    all_files = os.listdir(in_path)
    #print(all_files)
    for _ in range(int(ratio * len(all_files))):
        files = os.listdir(in_path)
        files.remove('test')
        files.remove('train')
        f = random.choice(files)
        os.rename(os.path.join(in_path, f), os.path.join(in_path, "test", f))
    files = os.listdir(in_path)
    files.remove('test')
    files.remove('train')
    for file in files:
        os.rename(os.path.join(in_path, file), os.path.join(in_path, "train", file))
```

```python
# Prepairing data 
# converts real captchas to black and white saves in a output dir
# in output dir splits all captchas to 5 pieces with names like {char}_N.jpg
# puts random *ratio percent of chars to test folder, other to train folder
# meta.json file shoild be added manually
in_dir = '../captcha-tensorflow-master/images/real_test/'
out_dir = '../captcha-tensorflow-master/images/real_test_bw/'
dir_to_bw(in_dir, out_dir, number=None, w=190)
path_cut_to_5(out_dir)
dir_to_bw(out_dir, out_dir)
#train_test_split('/home/andrey/1.RDV_Nanterre/Local/captcha-tensorflow-master/images/real_test_bw/')
```
